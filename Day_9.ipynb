```python
# break statement
count= 1
while count < 10:
  if count==3:
    break
  print(count)
  count +=1
print ("Loop Ended")
```

```python
# continue statement
count= 0
while count < 10:
  count +=1
  if count==3:
    continue
  print(count)
print ("Loop Ended")
```

```python
#pass statement
count = 0
while count <10:
  count +=1
  if count == 3:
   pass
  print(count)
print("Loop Ended")
```

```python
# ATM Menu System

balance = 100

while True:
    print("ATM MENU")
    print("1. Check Balance")
    print("2. Deposit")
    print("3. Exit")

    choice = int(input("Enter your choice: "))
# Check balance
    if choice == 1:
        print("Current Balance: Rs", balance)

#Deposit money
    elif choice == 2:
        amount = int(input("Enter amount to deposit: "))
        balance += amount
        print("Deposit Successful")
        print("Updated Balance: Rs", balance)

#Exit
    elif choice == 3:
        print("Thank you for using the ATM.")

    else:
        print("Invalid Choice! Please try again.")
        break
```
