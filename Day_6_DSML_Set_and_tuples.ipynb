```python
#intersection(&)
A = {1,2,3,4,5,6}
B = {3,4,5,6,7,8}
print(A & B)
```

```python
#difference(-)
A = {1,2,3,4,5,6}
B = {3,4,5,6,7,8}
print(A-B)
```
