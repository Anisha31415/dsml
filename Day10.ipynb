```python
def calculator(a, b, op):
   if op == "+":
      return a + b
   elif op == "-":
      return a-b
   elif op == "*":
      return a * b
   elif op == "/":
      return a / b
   else:
      return "Error"



```

```python
one = float(input("Enter a number"))
two = float(input("Enter another number"))
choice= (input("Enter operator like( +-*/)"))
result = calculator(one, two, choice)
print("Result:", result)

```

```python
def lst_operation():
  number =[]
  for i in range(5):
   value = input(f"Enter a value{i+1}:")
   if "." in value:
     if len(number) > 0:
      removed = number.pop()
      print(f"Float detected. removed:{removed}")
     else:
      print("float detected,but the list  is empty")
   else:
    number.append(int(value))
    print(f"Integer {value} added to the list")
  print("updated list:", number)




```

```python
lst_operation()
```
